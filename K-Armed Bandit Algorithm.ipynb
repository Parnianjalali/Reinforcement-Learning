```python
for i in range(2000):
    mu, sigma = 0, 1
    L=10
    q=np.random.normal(mu,sigma,L)
    np.reshape(q,(1,L))
    Q=np.zeros((1,L))
    N=np.zeros((1,L))
    for k in range(L):
        mu=q[k]
        R=np.random.normal(mu,sigma,1)
        N[:,k]=N[:,k]+1
        Q[:,k]=Q[:,k]+(1/N[:,k])*(R-Q[:,k])
    for x in range(1000):
        L=Q+2*np.sqrt(np.log(x)/N)
        t=np.argmax(L)
        mu=q[t]
        R=np.random.normal(mu,sigma,1)
        C[:,x]=R+C[:,x]
        N[:,t]=N[:,t]+1
        Q[:,t]=Q[:,t]+(1/N[:,t])*(R-Q[:,t])
```

```python
B=B/2000      
C=C/2000
x=np.arange(1,1001,step=1)
x=np.reshape(x,(1,1000))
B=np.reshape(B,(1000))
C=np.reshape(C,(1000))
V=np.reshape(x,(1000))
plt.plot(V,B,label='=0.1')
plt.plot(V,C,label='UCB C=2')
plt.xlabel('Steps')
plt.ylabel('Average reward')
plt.legend()
plt.show()     
```
